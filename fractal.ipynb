```python
#вариант 6
import numpy as np
import holoviews as hv
from holoviews import opts
from holoviews.streams import RangeXY
from numba import jit, prange
from IPython.display import Image


renderer = hv.renderer('bokeh')
hv.notebook_extension('bokeh')

@jit(nopython=True)
def julia(x, y, max_iters):
    """
    Вычисляет значение для множества Жюлиа для заданных координат x и y.

    Args:
        x: Координата по оси x.
        y: Координата по оси y.
        max_iters: Максимальное количество итераций.

    Returns:
        Количество итераций, на которое потребовалось, чтобы значение выйти за пределы множества.
    """
    c = complex(-0.8, 0.156)
    z = complex(x, y)
    for i in range(max_iters):
        z = z**4 + c
        if (z.real*z.real + z.imag*z.imag) >= 4:
            return i
    return 800

@jit(nopython=True, parallel=True)
def julia_set(min_x, max_x, min_y, max_y, image, iters):
    """
    Генерирует изображение множества Жюлиа.

    Args:
        min_x: Минимальное значение по оси x.
        max_x: Максимальное значение по оси x.
        min_y: Минимальное значение по оси y.
        max_y: Максимальное значение по оси y.
        image: Массив для хранения изображения.
        iters: Максимальное количество итераций.

    Returns:
        Массив с изображением множества Жюлиа.
    """
    height = image.shape[0]
    width = image.shape[1]

    pixel_size_x = (max_x - min_x) / width
    pixel_size_y = (max_y - min_y) / height

    for x in prange(width):
        real = min_x + x * pixel_size_x
        for y in prange(height):
            imag = min_y + y * pixel_size_y
            color = julia(real, imag, iters)
            image[y, x] = color

    return image

def plot_julia(x_range, y_range):
    """
    Создает и возвращает изображение множества Жюлиа для заданных диапазонов.

    Args:
        x_range : Диапазон значений по оси x.
        y_range : Диапазон значений по оси y.

    Returns:
        holoviews.Image: Изображение множества Жюлиа.
    """
    (x0, x1), (y0, y1) = x_range, y_range
    image = np.zeros((600, 600), dtype=np.uint8)
    return hv.Image(julia_set(x0, x1, -y1, -y0, image, 200),
                    bounds=(x0, y0, x1, y1))

range_stream = RangeXY(x_range=(-1., 1.), y_range=(-1., 1.))

dmap = hv.DynamicMap(plot_julia, streams=[range_stream])#.hist(log=True) работает не стабильно в разных средах разработки

dmap.opts(
    opts.Histogram(framewise=True, logy=True, width=200, xlim=(1, None)),
    opts.Image(cmap='fire', logz=True, height=600, width=600,
               xaxis=None, yaxis=None))
```
